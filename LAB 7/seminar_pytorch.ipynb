# Text processing

First we need next to collect a "vocabulary" of all unique tokens i.e. unique characters. We can then encode inputs as a sequence of character ids.

```python
#all unique characters go here
tokens = sorted(set(''.join(lines)))

tokens = list(tokens)

num_tokens = len(tokens)
print ('num_tokens = ', num_tokens)

assert 50 < num_tokens < 60, "Names should contain within 50 and 60 unique tokens depending on encoding"
```

### Convert characters to integers

Torch is built for crunching numbers, not strings.
To train our neural network, we'll need to replace characters with their indices in tokens list.

Let's compose a dictionary that does this mapping.

```python
token_to_id = {token: idx for idx, token in enumerate(tokens)}
```

```python
assert len(tokens) == len(token_to_id), "dictionaries must have same size"

for i in range(num_tokens):
    assert token_to_id[tokens[i]] == i, "token identifier must be it's position in tokens list"

print("Seems alright!")
```

```python
def to_matrix(lines, max_len=None, pad=token_to_id[' '], dtype='int32', batch_first = True):
    """Casts a list of names into rnn-digestable matrix"""

    max_len = max_len or max(map(len, lines))
    lines_ix = np.zeros([len(lines), max_len], dtype) + pad

    for i in range(len(lines)):
        line_ix = [token_to_id[c] for c in lines[i]]
        lines_ix[i, :len(line_ix)] = line_ix

    if not batch_first: # convert [batch, time] into [time, batch]
        lines_ix = np.transpose(lines_ix)

    return lines_ix
```

# Recurrent neural network

We can rewrite recurrent neural network as a consecutive application of dense layer to input $x_t$ and previous rnn state $h_t$. This is exactly what we're gonna do now.
<img src="https://github.com/yandexdataschool/Practical_RL/blob/master/week07_%5Brecap%5D_rnn/rnn.png?raw=1" width=480>

Since we're training a language model, there should also be:
* An embedding layer that converts character id x_t to a vector.
* An output layer that predicts probabilities of next phoneme

```python
import torch, torch.nn as nn
import torch.nn.functional as F

class CharRNNCell(nn.Module):
    """
    Implement the scheme above as torch module
    """
    def __init__(self, num_tokens=len(tokens), embedding_size=16, rnn_num_units=64):
        super(self.__class__,self).__init__()
        self.num_units = rnn_num_units

        self.embedding = nn.Embedding(num_tokens, embedding_size)
        self.rnn_update = nn.Linear(embedding_size + rnn_num_units, rnn_num_units)
        self.rnn_to_logits = nn.Linear(rnn_num_units, num_tokens)

    def forward(self, x, h_prev):
        """
        This method computes h_next(x, h_prev) and log P(x_next | h_next)
        We'll call it repeatedly to produce the whole sequence.

        :param x: batch of character ids, int64[batch_size]
        :param h_prev: previous rnn hidden states, float32 matrix [batch, rnn_num_units]
        """
        # get vector embedding of x
        x_emb = self.embedding(x)

        # compute next hidden state using self.rnn_update
        # hint: use torch.cat(..., dim=...) for concatenation
        h_next = self.rnn_update(torch.cat([x_emb, h_prev], dim=1))

        h_next = torch.tanh(h_next)

        assert h_next.size() == h_prev.size()

        #compute logits for next character probs
        logits = self.rnn_to_logits(h_next)

        return h_next, F.log_softmax(logits, -1)

    def initial_state(self, batch_size):
        """ return rnn state before it processes first input (aka h0) """
        return torch.zeros(batch_size, self.num_units)
```

```python
char_rnn = CharRNNCell()
```

### RNN loop

Once we've defined a single RNN step, we can apply it in a loop to get predictions on each step.

```python
def rnn_loop(char_rnn, batch_ix):
    """
    Computes log P(next_character) for all time-steps in lines_ix
    :param lines_ix: an int32 matrix of shape [batch, time], output of to_matrix(lines)
    """
    batch_size, max_length = batch_ix.size()
    hid_state = char_rnn.initial_state(batch_size)
    logprobs = []

    for x_t in batch_ix.transpose(0,1):
        hid_state, logp_next = char_rnn(x_t, hid_state)  # <-- here we call your one-step code
        logprobs.append(logp_next)

    return torch.stack(logprobs, dim=1)
```

```python
batch_ix = to_matrix(lines[:5])
batch_ix = torch.tensor(batch_ix, dtype=torch.int64)

logp_seq = rnn_loop(char_rnn, batch_ix)

assert torch.max(logp_seq).data.numpy() <= 0
assert tuple(logp_seq.size()) ==  batch_ix.shape + (num_tokens,)
```

### Likelihood and gradients

We can now train our neural network to minimize crossentropy (maximize log-likelihood) with the actual next tokens.

To do so in a vectorized manner, we take `batch_ix[:, 1:]` - a matrix of token ids shifted i step to the left so i-th element is acutally the "next token" for i-th prediction

```python
predictions_logp = logp_seq[:, :-1]
actual_next_tokens = batch_ix[:, 1:]

logp_next = torch.gather(predictions_logp, dim=2, index=actual_next_tokens[:,:,None])

loss = -logp_next.mean()
```

```python
loss.backward()
```

```python
for w in char_rnn.parameters():
    assert w.grad is not None and torch.max(torch.abs(w.grad)).data.numpy() != 0, \
        "Loss is not differentiable w.r.t. a weight with shape %s. Check forward method." % (w.size(),)
```

### The training loop

We train our char-rnn exactly the same way we train any deep learning model: by minibatch sgd.

The only difference is that this time we sample strings, not images or sound.

```python
from IPython.display import clear_output
from random import sample

char_rnn = CharRNNCell()
opt = torch.optim.Adam(char_rnn.parameters())
history = []
```

```python

for i in range(1000):
    batch_ix = to_matrix(sample(lines, 32), max_len=MAX_LENGTH)
    batch_ix = torch.tensor(batch_ix, dtype=torch.int64)

    logp_seq = rnn_loop(char_rnn, batch_ix)

    # compute loss
    predictions_logp = logp_seq[:, :-1]
    actual_next_tokens = batch_ix[:, 1:]
    logp_next = torch.gather(predictions_logp, dim=2, index=actual_next_tokens[:, :, None])

    loss = -logp_next.mean()


    # train with backprop
    opt.zero_grad()
    loss.backward()
    opt.step()

    history.append(loss.data.numpy())
    if (i+1)%100==0:
        clear_output(True)
        plt.plot(history,label='loss')
        plt.legend()
        plt.show()

assert np.mean(history[:10]) > np.mean(history[-10:]), "RNN didn't converge."
```

### RNN: sampling
Once we've trained our network a bit, let's get to actually generating stuff.
All we need is the single rnn step function you have defined in `char_rnn.forward`.

```python
def generate_sample(char_rnn, seed_phrase=' ', max_length=MAX_LENGTH, temperature=1.0):
    '''
    The function generates text given a phrase of length at least SEQ_LENGTH.
    :param seed_phrase: prefix characters. The RNN is asked to continue the phrase
    :param max_length: maximum output length, including seed_phrase
    :param temperature: coefficient for sampling.  higher temperature produces more chaotic outputs,
                        smaller temperature converges to the single most likely output
    '''

    x_sequence = [token_to_id[token] for token in seed_phrase]
    x_sequence = torch.tensor([x_sequence], dtype=torch.int64)
    hid_state = char_rnn.initial_state(batch_size=1)

    #feed the seed phrase, if any
    for i in range(len(seed_phrase) - 1):
        hid_state, _ = char_rnn(x_sequence[:, i], hid_state)

    #start generating
    for _ in range(max_length - len(seed_phrase)):
        hid_state, logp_next = char_rnn(x_sequence[:, -1], hid_state)
        p_next = F.softmax(logp_next / temperature, dim=-1).data.numpy()[0]

        # sample next token and push it back into x_sequence
        next_ix = np.random.choice(num_tokens,p=p_next)
        next_ix = torch.tensor([[next_ix]], dtype=torch.int64)
        x_sequence = torch.cat([x_sequence, next_ix], dim=1)

    return ''.join([tokens[ix] for ix in x_sequence.data.numpy()[0]])
```

### More seriously

What we just did is a manual low-level implementation of RNN. While it's cool, i guess you won't like the idea of re-writing it from scratch on every occasion.

As you might have guessed, torch has a solution for this. To be more specific, there are two options:
* `nn.RNNCell(emb_size, rnn_num_units)` - implements a single step of RNN just like you did. Basically concat-linear-tanh
* `nn.RNN(emb_size, rnn_num_units` - implements the whole rnn_loop for you.

There's also `nn.LSTMCell` vs `nn.LSTM`, `nn.GRUCell` vs `nn.GRU`, etc. etc.

In this example we'll rewrite the char_rnn and rnn_loop using high-level rnn API.

```python
class CharRNNLoop(nn.Module):
    def __init__(self, num_tokens=num_tokens, emb_size=16, rnn_num_units=64):
        super(self.__class__, self).__init__()
        self.emb = nn.Embedding(num_tokens, emb_size)
        self.rnn = nn.RNN(emb_size, rnn_num_units, batch_first=True)
        self.hid_to_logits = nn.Linear(rnn_num_units, num_tokens)

    def forward(self, x):
        h_seq, _ = self.rnn(self.emb(x))
        next_logits = self.hid_to_logits(h_seq)
        next_logp = F.log_softmax(next_logits, dim=-1)
        return next_logp

model = CharRNNLoop()
```

```python
# the model applies over the whole sequence
batch_ix = to_matrix(sample(lines, 32), max_len=MAX_LENGTH)
batch_ix = torch.tensor(batch_ix, dtype=torch.int64)

logp_seq = model(batch_ix)

# compute loss. This time we use nll_loss with some duct tape
loss = F.nll_loss(logp_seq[:, :-1].contiguous().view(-1, num_tokens),
                  batch_ix[:, 1:].contiguous().view(-1))

loss.backward()
```

Here's another example

```python
import torch, torch.nn as nn
import torch.nn.functional as F

class CharLSTMCell(nn.Module):
    """
    Implements something like CharRNNCell, but with LSTM
    """
    def __init__(self, num_tokens=len(tokens), embedding_size=16, rnn_num_units=64):
        super(self.__class__,self).__init__()
        self.num_units = rnn_num_units
        self.emb = nn.Embedding(num_tokens, embedding_size)
        self.lstm = nn.LSTMCell(embedding_size, rnn_num_units)
        self.rnn_to_logits = nn.Linear(rnn_num_units, num_tokens)

    def forward(self, x, prev_state):
        (prev_h, prev_c) = prev_state
        (next_h, next_c) = self.lstm(self.emb(x), (prev_h, prev_c))
        logits = self.rnn_to_logits(next_h)

        return (next_h, next_c), F.log_softmax(logits, -1)

    def initial_state(self, batch_size):
        """ LSTM has two state variables, cell and hid """
        return torch.zeros(batch_size, self.num_units), torch.zeros(batch_size, self.num_units)

char_lstm = CharLSTMCell()
```

```python
# the model applies over the whole sequence
batch_ix = to_matrix(sample(lines, 32), max_len=MAX_LENGTH)
batch_ix = torch.tensor(batch_ix, dtype=torch.int64)

logp_seq = rnn_loop(char_lstm, batch_ix)

# compute loss. This time we use nll_loss with some duct tape
loss = F.nll_loss(logp_seq[:, :-1].contiguous().view(-1, num_tokens),
                  batch_ix[:, 1:].contiguous().view(-1))

loss.backward()
```

__Bonus quest: __ implement a model that uses 2 LSTM layers (the second lstm uses the first as input) and train it on your data.

```python
class TwoLayerLSTM(nn.Module):
    """
    Implements a 2-layer LSTM for character-level language modeling
    The second LSTM layer takes the output of the first LSTM as input
    """
    def __init__(self, num_tokens=len(tokens), embedding_size=16, rnn_num_units=64):
        super(TwoLayerLSTM, self).__init__()
        self.num_units = rnn_num_units

        self.emb = nn.Embedding(num_tokens, embedding_size)

        self.lstm1 = nn.LSTMCell(embedding_size, rnn_num_units)

        self.lstm2 = nn.LSTMCell(rnn_num_units, rnn_num_units)

        self.rnn_to_logits = nn.Linear(rnn_num_units, num_tokens)

    def forward(self, x, prev_states):
        """
        One step forward through both LSTM layers

        :param x: batch of character ids, int64[batch_size]
        :param prev_states: tuple of ((h1, c1), (h2, c2)) for both layers
        :return: new_states, log_probs
        """
        (prev_h1, prev_c1), (prev_h2, prev_c2) = prev_states

        x_emb = self.emb(x)
        next_h1, next_c1 = self.lstm1(x_emb, (prev_h1, prev_c1))

        next_h2, next_c2 = self.lstm2(next_h1, (prev_h2, prev_c2))

        logits = self.rnn_to_logits(next_h2)
        log_probs = F.log_softmax(logits, dim=-1)

        return (next_h1, next_c1), (next_h2, next_c2), log_probs

    def initial_state(self, batch_size):
        """Return initial states for both LSTM layers"""
        h1 = torch.zeros(batch_size, self.num_units)
        c1 = torch.zeros(batch_size, self.num_units)
        h2 = torch.zeros(batch_size, self.num_units)
        c2 = torch.zeros(batch_size, self.num_units)
        return ((h1, c1), (h2, c2))
```

```python
def rnn_loop_two_layer(model, batch_ix):
    """
    Computes log P(next_character) for all time-steps using 2-layer LSTM

    :param model: TwoLayerLSTM model
    :param batch_ix: an int32 matrix of shape [batch, time], output of to_matrix(lines)
    """
    batch_size, max_length = batch_ix.size()
    states = model.initial_state(batch_size)
    logprobs = []

    for x_t in batch_ix.transpose(0, 1):
        states_layer1, states_layer2, logp_next = model(x_t, states)
        states = (states_layer1, states_layer2)
        logprobs.append(logp_next)

    return torch.stack(logprobs, dim=1)
```

```python
def generate_sample_two_layer(model, seed_phrase=' ', max_length=MAX_LENGTH, temperature=0.8):
    """
    Generate text using the two-layer LSTM model

    :param model: TwoLayerLSTM model
    :param seed_phrase: prefix characters
    :param max_length: maximum output length
    :param temperature: sampling temperature (lower = more conservative, higher = more diverse)
    """
    model.eval()

    with torch.no_grad():
        x_sequence = [token_to_id.get(token, token_to_id[' ']) for token in seed_phrase]
        x_sequence = torch.tensor([x_sequence], dtype=torch.int64)

        states = model.initial_state(batch_size=1)

        for i in range(len(seed_phrase) - 1):
            states_layer1, states_layer2, _ = model(x_sequence[:, i], states)
            states = (states_layer1, states_layer2)

        for _ in range(max_length - len(seed_phrase)):
            states_layer1, states_layer2, logp_next = model(x_sequence[:, -1], states)
            states = (states_layer1, states_layer2)

            p_next = F.softmax(logp_next / temperature, dim=-1).cpu().numpy()[0]
            next_ix = np.random.choice(num_tokens, p=p_next)

            next_ix_tensor = torch.tensor([[next_ix]], dtype=torch.int64)
            x_sequence = torch.cat([x_sequence, next_ix_tensor], dim=1)

        result = ''.join([tokens[ix] for ix in x_sequence.cpu().numpy()[0]])

    model.train()
    return result
```

```python
from IPython.display import clear_output
from random import sample

two_layer_lstm = TwoLayerLSTM(
    num_tokens=num_tokens,
    embedding_size=16,
    rnn_num_units=64
)

optimizer = torch.optim.Adam(two_layer_lstm.parameters(), lr=0.003)
history = []

print("Training 2-layer LSTM...")
print(f"Number of parameters: {sum(p.numel() for p in two_layer_lstm.parameters())}")

for i in range(1500):
    batch_ix = to_matrix(sample(lines, 32), max_len=MAX_LENGTH)
    batch_ix = torch.tensor(batch_ix, dtype=torch.int64)

    optimizer.zero_grad()

    logp_seq = rnn_loop_two_layer(two_layer_lstm, batch_ix)

    loss = F.nll_loss(
        logp_seq[:, :-1].contiguous().view(-1, num_tokens),
        batch_ix[:, 1:].contiguous().view(-1)
    )

    loss.backward()

    torch.nn.utils.clip_grad_norm_(two_layer_lstm.parameters(), max_norm=1.0)

    optimizer.step()

    history.append(loss.item())

    if (i + 1) % 100 == 0:
        clear_output(True)
        plt.figure(figsize=(12, 4))

        plt.subplot(1, 2, 1)
        plt.plot(history, label='Loss', alpha=0.7)
        plt.plot(range(0, len(history), 100),
                [np.mean(history[i:i+100]) for i in range(0, len(history), 100)],
                'r-', linewidth=2, label='100-step average')
        plt.xlabel('Iteration')
        plt.ylabel('Loss')
        plt.title('Training Loss')
        plt.legend()
        plt.grid(True, alpha=0.3)

        plt.subplot(1, 2, 2)
        examples = []
        for _ in range(3):
            sample_name = generate_sample_two_layer(two_layer_lstm, temperature=0.8)
            examples.append(sample_name)

        plt.text(0.1, 0.5, '\n'.join(examples), fontsize=12,
                verticalalignment='center', family='monospace')
        plt.axis('off')
        plt.title('Generated Names')

        plt.tight_layout()
        plt.show()

        print(f"Iteration {i+1}, Loss: {loss.item():.4f}")

print("Training complete!")
```

```python
print("Testing 2-Layer LSTM Model")

for temp in [0.5, 0.8, 1.2]:
    print(f"\nTemperature = {temp}:")
    print("-" * 30)
    for _ in range(5):
        generated = generate_sample_two_layer(two_layer_lstm, seed_phrase=' ', temperature=temp)
        print(f"  {generated}")

print("Generating with different seed phrases:")

for seed in [' ', ' A', ' El', ' Joh']:
    print(f"\nSeed: '{seed}'")
    print("-" * 30)
    for _ in range(3):
        generated = generate_sample_two_layer(two_layer_lstm, seed_phrase=seed, temperature=0.8)
        print(f"  {generated}")
```
